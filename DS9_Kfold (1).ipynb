```
#defining the range of the box plots by calculating the inter quartile range (IQR)
Q1 = df['SepalWidthCm'].quantile(0.25)
Q3 = df['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1    

sepal_width_new = (df['SepalWidthCm'] >= Q1 - 1.5 * IQR) & (df['SepalWidthCm'] <= Q3 + 1.5 *IQR)


sepal_width_new
```

```
# fitting the model 
from sklearn.naive_bayes import GaussianNB
X_train, X_test, Y_train, Y_test =train_test_split(features,target_variable , test_size=0.5, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, Y_train).predict(X_test)
```

```
from sklearn.metrics import classification_report

target_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
print(classification_report(Y_test, y_pred, target_names=target_names))
```

```
from sklearn import metrics 
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(Y_test, y_pred)*100)
```

```
#introducing Kfold
from sklearn.model_selection import KFold,cross_val_score
k_fold = KFold(n_splits=15)
cv_results = cross_val_score(GaussianNB(), X_train, Y_train, cv=k_fold)
print("%s: %f ; %s: %f" % ('Naive Bayes with Kfold accuracy %', cv_results.mean()*100,'Standard deviation %',cv_results.std()*100))
```
